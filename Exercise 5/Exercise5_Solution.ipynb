## Question 1a

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree

# Load the dataset
data = pd.read_csv("/Users/aakashkandasamy/Documents/NTU/NTU COURSES SEM 2/SC1015/train.csv", header=0)

# Create Train and Test datasets
X = data[["SalePrice"]]
y = data["CentralAir"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit Classification Tree model with max depth 2 on Train set
tree2 = DecisionTreeClassifier(max_depth=2)
tree2.fit(X_train, y_train)

# Print the tree
plt.figure(figsize=(10,6))
plot_tree(tree2, feature_names=["SalePrice"], class_names=["No", "Yes"], filled=True)
plt.show()

# Fit Classification Tree model with max depth 4 on Train set
tree4 = DecisionTreeClassifier(max_depth=4)
tree4.fit(X_train, y_train)

# Print the tree
plt.figure(figsize=(15,10))
plot_tree(tree4, feature_names=["SalePrice"], class_names=["No", "Yes"], filled=True)
plt.show()

y_train_pred = tree2.predict(X_train)
y_test_pred = tree2.predict(X_test)

print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy for Depth 2 \t:", tree2.score(X_train, y_train))
print()

print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy for Depth 2 \t:", tree2.score(X_test, y_test))
print()

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])


y_train_pred = tree4.predict(X_train)
y_test_pred = tree4.predict(X_test)

print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy for Depth 4 \t:", tree4.score(X_train, y_train))
print()

print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy for Depth 4 \t:", tree4.score(X_test, y_test))
print()


# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])
```
