# Question 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv("/Users/aakashkandasamy/Documents/NTU/NTU COURSES SEM 2/SC1015/train.csv", header=0)

train_data, test_data, train_target, test_target = train_test_split(data[['GrLivArea']], data['SalePrice'], test_size=0.2)

reg = LinearRegression().fit(train_data, train_target)

print("Coefficients: ", reg.coef_)
print("Intercept: ", reg.intercept_)

plt.scatter(train_data, train_target, color='blue')
plt.plot(train_data, reg.predict(train_data), color='red', linewidth=2)
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.show()

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

train_predictions = reg.predict(train_data)

r2_train = r2_score(train_target, train_predictions)
mse_train = mean_squared_error(train_target, train_predictions)

print("Train R-squared: ", r2_train)
print("Train MSE: ", mse_train)

test_predictions = reg.predict(test_data)

mae_test = mean_absolute_error(test_target, test_predictions)
mse_test = mean_squared_error(test_target, test_predictions)

print("Test MAE: ", mae_test)
print("Test MSE: ", mse_test)


```

# Question 3

```python
threshold = 3

outliers_gla = data[np.abs(data["GrLivArea"] - data["GrLivArea"].mean()) > (threshold * data["GrLivArea"].std())]

outliers_sp = data[np.abs(data["SalePrice"] - data["SalePrice"].mean()) > (threshold * data["SalePrice"].std())]

outliers = set(outliers_gla.index.tolist() + outliers_sp.index.tolist())

data1 = data[~data.index.isin(outliers)].reset_index(drop=True)
data1.info()
```

## The data has been cleaned now with the removal of the outliers. A threshold score of 3 was used to determine the values that are 3 sd away from the mean to be classified as outliers. 

```python
train_data3, test_data3, train_target3, test_target3 = train_test_split(data1[['GrLivArea']], data1['SalePrice'], test_size=0.2)

reg3 = LinearRegression().fit(train_data3, train_target3)

print("Coefficients: ", reg3.coef_)
print("Intercept: ", reg3.intercept_)

plt.scatter(train_data3, train_target3, color='blue')
plt.plot(train_data3, reg3.predict(train_data3), color='red', linewidth=2)
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.show()

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

train_predictions3 = reg3.predict(train_data3)

r2_train3 = r2_score(train_target3, train_predictions3)
mse_train3 = mean_squared_error(train_target3, train_predictions3)

print("Train R-squared: ", r2_train3)
print("Train MSE: ", mse_train3)

test_predictions3 = reg3.predict(test_data3)

mae_test3 = mean_absolute_error(test_target3, test_predictions3)
mse_test3 = mean_squared_error(test_target3, test_predictions3)

print("Test MAE: ", mae_test3)
print("Test MSE: ", mse_test3)
```

```python

def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers1 = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers1
```

```python
numeric_data = pd.DataFrame(data[['GrLivArea', 'SalePrice']])
numeric_data.describe()

for var in numeric_data:
    outliers2 = find_outliers_IQR(numeric_data[var])
    print(outliers2)
    
data2 = data[~data.index.isin(outliers1)].reset_index(drop=True)
data2.info()
```
