### Generic Loop to Scrape

It turns out that Wikipedia has medal table at index `0` for years from `1896` to `1988`, and at index `1` for years from `1992` to `2016`. So, we will need to indicate the table matching "*Rank*" to scrape the correct table all the time. Also note that Summer Olympics were not held in `1916`, `1940`, `1944`, due to the world wars. So, we will omit those years manually. Let's get only TOP 10 this time, for every year Summer Olympics took place.

```python
first_year = 1896
last_year = 2016
omit_years = [1916, 1940, 1944]

for year in range(first_year, last_year + 1, 4):
    # skip the years olympics did not take place
    if year in omit_years:
        continue

    # fetch the html page with the url for year
    wiki_url = "https://en.wikipedia.org/wiki/" + str(year) + "_Summer_Olympics_medal_table"
    
    # scrape table(s) only if it matches "Rank"
    medal_html = pd.read_html(wiki_url, match = "Rank")
    medalTable = medal_html[0]
    
    # extract only TOP 10 countries in the table
    medalData = medalTable.iloc[:10]
    
    # display the year and the top 10 countries
    print()
    print("================================================")
    print(year)
    display(medalData)
```

If you understood the logic so far, it's only a matter of time till you can do the following with the tables.

```python
first_year = 1896
last_year = 2016
omit_years = [1916, 1940, 1944]

medalDataFrames = []

for year in range(first_year, last_year + 1, 4):
    # skip the years olympics did not take place
    if year in omit_years:
        continue

    # fetch the html page with the url for year
    wiki_url = "https://en.wikipedia.org/wiki/" + str(year) + "_Summer_Olympics_medal_table"
    
    # scrape table(s) only if it matches "Rank"
    medal_html = pd.read_html(wiki_url, match = "Rank")
    medalTable = medal_html[0]
    
    # extract only TOP 10 countries in the table
    medalData = medalTable.iloc[:10]
    
    # rename the columns to make tables uniform
    medalData.columns = ["Rank", "Nation", "Gold", "Silver", "Bronze", "Total"]
    
    # store the medal dataframe in the main list
    medalDataFrames.append(medalData)
    
medalDataAll = pd.concat(medalDataFrames, axis = 1)
```

Now that you've got all the data for TOP 10 teams from `1896` to `2016`, go ahead and do whatever you want!

```python
# e.g., Nations across the years in TOP 10 list
medalDataAll.Nation
```

## Extra Problem : Just for fun!

Any structured website can be scraped for tables in the same way. However, what would you do for data that are not in a table format? Can you extract the name of the movie, its rating and genres from https://www.imdb.com/title/tt0441773/ using some other library in python? Give it a shot! :-)

```python
# first, you will need a GET REQUEST for the HTML
from requests import get
url = 'https://www.imdb.com/title/tt0441773/'
response = get(url)

# then, you will need Beautiful Soup for the data
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# once you got the html_soup, extract the title_block
title_block = html_soup.find_all('div', class_ = 'title_block')[0]
title_block
```

```python
# now you can extract targeted data from this mess
print("Rating : ", title_block.div.strong.span.text)
print("Title  : ", title_block.h1.text)
print("Year   : ", title_block.h1.span.a.text)
```

```python
# extracting multiple occurrences is a little tricky
genres = title_block.find('div', class_ = 'subtext')
for x in genres.find_all('a')[:3]:
    print(x.text)
```
